```python
def is_float(x):
    try:
        float(x)
    except:
        return False
    return True
```

```python
df2=df1.copy()
df2[~df2["total_sqft"].apply(is_float)]
```

```python
def convert_to_num(x):
    token=x.split("-")
    if len(token)==2:
        return (float(token[0])+float(token[1]))//2
    try:
        return float(x)
    except:
        return None
```

```python
df3=df2.copy()
df3["total_sqft"]=df3["total_sqft"].apply(convert_to_num)
```

```python
df4=df3.copy()
df4["Price per Squarefeet"]=df4["price"]*100000/df4["total_sqft"]
df4.head()
```

```python
df8=df7.copy()
#check The Price_per_SquareFeet
df8["Price per Squarefeet"].describe()
```

```python
import numpy as np
def Calculate(df):
    df_out=pd.DataFrame()
    for key,subdf in df.groupby('location'):
        mean=np.mean(subdf["Price per Squarefeet"])
        std=np.std(subdf["Price per Squarefeet"])
        red_df=subdf[(subdf["Price per Squarefeet"]>=(mean-std)) & (subdf["Price per Squarefeet"]<=(mean+std))]
        df_out=pd.concat([df_out,red_df],ignore_index=True)
    return df_out
df8=Calculate(df8)
        
```

```python
#Check that Two Bedrooms and Three Bedrooms Have the Prices Range
import matplotlib.pyplot as plt
def plotter(df,location):
    bhk_2=df[(df.location==location) & (df.bhk==2)]
    bhk_3=df[(df.location==location) & (df.bhk==3)]
    plt.rcParams['figure.figsize']=(15,10)
    plt.scatter(df.total_sqft,df.price,color="blue",label='2 BHK',s=50)
    plt.scatter(df.total_sqft,df.price,color="green",marker="+",label="3 BHK",s=50)
    plt.xlabel("Total_squarefeet")
    plt.ylabel("Price")
    plt.legend()
df9=df8.copy()
plotter(df9,"Balagere")
    
```

```python
#Remove The Bhk Outlier
def Remove_outlier(df):
    exclude_indeces=np.array([])
    for location,location_df in df.groupby('location'):
        bhk_states={}
        for bhk,bhk_df in df.groupby('bhk'):
            bhk_states[bhk]={"mean":np.mean(bhk_df["Price per Squarefeet"]),
                             "std":np.std(bhk_df["Price per Squarefeet"]),
                             "count":bhk_df.shape[0]}
        for bhk,bhk_df in location_df.groupby('bhk'):
            stats=bhk_states.get(bhk-1)
            if stats and stats["count"]>5:
                exclude_indeces=np.append(exclude_indeces,bhk_df[bhk_df["Price per Squarefeet"]<(stats["mean"])].index.values)
    return df.drop(exclude_indeces,axis="index")
            
```

```python
df9=df8.copy()
df9=Remove_outlier(df9)
```
